# Eval: the 5 assignment queries

Run each of the 5 required queries end-to-end and render the top-k image grid so results
can be sanity-checked visually. Requires the index to already be built
(`python -m src.indexer.build_index`) and, for LLM-based query parsing,
`GEMINI_API_KEY` exported in the environment.

```python
import sys
sys.path.insert(0, "..")

import matplotlib.pyplot as plt
from PIL import Image

from src.retriever.query import run_query
```

```python
EVAL_QUERIES = [
    "A person in a bright yellow raincoat.",
    "Professional business attire inside a modern office.",
    "Someone wearing a blue shirt sitting on a park bench.",
    "Casual weekend outfit for a city walk.",
    "A red tie and a white shirt in a formal setting.",
]
```

```python
def show_results(query_text: str, k: int = 5) -> None:
    results = run_query(query_text, k=k)
    fig, axes = plt.subplots(1, len(results), figsize=(4 * len(results), 4))
    if len(results) == 1:
        axes = [axes]
    fig.suptitle(query_text)
    for ax, r in zip(axes, results):
        ax.imshow(Image.open(r["image_path"]))
        ax.set_title(f"score={r['final_score']:.3f}", fontsize=9)
        ax.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
for q in EVAL_QUERIES:
    show_results(q)
```
